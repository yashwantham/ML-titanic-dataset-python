```python
def barchart(feature):
    survived = tab[tab['Survived']==1][feature].value_counts()
    dead = tab[tab['Survived']==0][feature].value_counts()
    x2 = pd.DataFrame([survived,dead])
    x2.index = ['Survived','Dead']
    x2.plot(kind='bar', figsize=(6,5))
    return x2
```
